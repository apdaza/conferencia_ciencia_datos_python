```python
import pandas as pd
import matplotlib.pyplot as plt
```

Cargamos la data

```python
df = pd.read_csv('us_states_covid19_daily.csv')
df.head()
```

```python
df.shape
```

```python
df_hi = df[df.positive >= 500000]
df_hi.head()
```

```python
df_hi.shape
```

```python
plt.figure(figsize=(8, 9))
plt.bar(df_hi['state'], df_hi['positive'], color='blue')
plt.title('Estados con los mayores casos de COVID-19')
plt.xlabel('Estados')
plt.ylabel('Casos en millones')
```

```python
df_FL = df[df.state == 'FL']
df_FL = df_FL.sort_values(by='date')
df_FL.head()
```

```python
df_CA = df[df.state == 'CA']
df_CA = df_CA.sort_values(by='date')
df_CA.head()
```

```python
df_TX = df[df.state == 'TX']
df_TX = df_TX.sort_values(by='date')
df_TX.head()
```

```python
plt.plot(df_FL['date'], df_FL['positive'], color="red")
plt.title('Casos positivos en Florida')
plt.ylabel('NÃºmero de casos')
plt.xlabel('Desde 03-04-2020 al 12-06-2020')
plt.xticks([])
plt.show()
```

```python
plt.plot(df_FL['date'], df_FL['death'], color="red")
plt.title('Muertes en Florida')
plt.ylabel('NÃºmero de casos')
plt.xlabel('Desde 03-04-2020 al 12-06-2020')
plt.xticks([])
plt.show()
```

```python
plt.plot(df_FL['date'], df_FL['positive'], color="red")
plt.plot(df_FL['date'], df_FL['death'], color="black")
plt.title('Casos positivos y muertes en Florida')
plt.ylabel('NÃºmero de casos')
plt.xlabel('Desde 03-04-2020 al 12-06-2020')
plt.xticks([])
plt.show()
```

```python
df_NY = df[df.state == 'NY']
df_NY = df_NY.sort_values(by='date')
df_NY.head()
```

```python
plt.plot(df_FL['date'], df_FL['positive'], color="red", label='FL')
plt.plot(df_CA['date'], df_CA['positive'], color="blue", label='CA')
plt.plot(df_TX['date'], df_TX['positive'], color="green", label="TX")
plt.plot(df_NY['date'], df_NY['positive'], color="black", label="NY")
plt.title('Casos positivos')
plt.ylabel('NÃºmero de casos')
plt.xlabel('Desde 03-04-2020 al 12-06-2020')
plt.legend()
plt.xticks([])
plt.show()
```

```python
plt.plot(df_FL['date'], df_FL['death'], color="red", label='FL')
plt.plot(df_CA['date'], df_CA['death'], color="blue", label='CA')
plt.plot(df_TX['date'], df_TX['death'], color="green", label="TX")
plt.plot(df_NY['date'], df_NY['death'], color="black", label="NY")
plt.title('Muertes por COVID')
plt.ylabel('NÃºmero de casos')
plt.xlabel('Desde 03-04-2020 al 12-06-2020')
plt.legend()
plt.xticks([])
plt.show()
```
